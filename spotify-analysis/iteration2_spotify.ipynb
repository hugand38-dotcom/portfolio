### Stocker conjointement l’année de sortie (par ordre croissant) et le nombre de chansons du top 100 dans un tuple. 

Les tuples seront stockés dans une liste (ex : [(1985, 1), (1986, 0), (1987, 0), ...])

Vous pouvez essayer avec 3 méthodes différentes: 
- Boucle for

```python
# Methode 1 : boucle for
nb_songs_per_year=[]
for y in range (1985, 2025):
    nb_songs_per_year.append((y, annee_sortie.count(y)))
print(nb_songs_per_year)
```

**PLUS DIFFICILE : Trier les tuples (Année de sortie, nombre de chansons) en fonction du nombre de chansons par ordre décroissant**

=> permet de retrouver l'année la plus productive en termes de Top Song


```python
# tri par nombre de chansons
annee_tri_tuples = sorted(nb_songs_per_year, key = lambda x: x[1], reverse = True)
print(annee_tri_tuples)
```

### Identifier tous les artistes uniques (en séparant les collaborations)

=> Fonctions split() et strip() : https://www.geeksforgeeks.org/python/difference-between-strip-and-split-in-python/

```python
# Chaque entrée peut contenir plusieurs artistes séparés par '&'
artiste=[]
for y in data[1:] :
        artiste.append(y[2])
print(artiste)

artistes_uniques = []
for y in data[1:]:
        artistes = y[2].split('&')
        for artiste in artistes:
                artiste = artiste.strip()
                if artiste not in artistes_uniques:
                        artistes_uniques.append(artiste)
print(artistes_uniques)
```

```python
# Affichage de tous les artistes, par ordre alphabétique
print(sorted(artistes_uniques))
```

### Compter le nombre de chansons par artiste avec un Dict

Vous pouvez utiliser deux méthodes différentes:
- Une boucle
- Un compteur Counter : https://www.geeksforgeeks.org/python/counters-in-python-set-1/


```python
# Méthode 1 : boucle
chanson_par_artiste = {}
for y in data[1:]:
    artistes = y[2]
    if artistes in chanson_par_artiste:
        chanson_par_artiste[artistes] += 1
    else:
        chanson_par_artiste[artistes] = 1
chanson_par_artiste
```

### Artiste le plus prolifique


```python
artiste_max = max(chanson_par_artiste, key=lambda x: chanson_par_artiste[x])
print(artiste_max)
```
